```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
n = 1000
departments = ['HR', 'IT', 'Finance', 'Marketing', 'Operations', 'Sales']
data = {
    'Employee_ID': range(1, n+1),
    'Age': np.random.randint(21, 60, n),
    'Department': np.random.choice(departments, n),
    'Years_Experience': np.random.randint(0, 35, n),
    'Training_Hours': np.random.randint(5, 100, n),
    'Performance_Score': np.round(np.random.uniform(2.0, 5.0, n), 2),
    'Salary': np.random.randint(30000, 150000, n),
    'Work_Life_Balance': np.random.randint(1, 6, n)  # 1 (poor) to 5 (excellent)
}
employee_df = pd.DataFrame(data)
employee_df.head()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Department', data=employee_df)
plt.title('Employee Distribution per Department')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Department', y='Salary', data=employee_df, estimator=np.mean)
plt.title('Average Salary per Department')
plt.ylabel('Average Salary')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Department', y='Training_Hours', data=employee_df, estimator=np.mean)
plt.title('Average Training Hours by Department')
plt.ylabel('Average Training Hours')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Training_Hours', y='Performance_Score', hue='Department', data=employee_df)
plt.title('Training Hours vs Performance Score')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Department', y='Work_Life_Balance', data=employee_df)
plt.title('Department-wise Work-Life Balance')
plt.ylabel('WLB Score (15)')
plt.show()
```
